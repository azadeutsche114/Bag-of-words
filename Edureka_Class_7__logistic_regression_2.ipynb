# Logistic Regression

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X = cancer.data # input
y = cancer.target # output
```

```python
print(cancer.DESCR)
```

```python
import pandas as pd
df = pd.DataFrame(X, columns=cancer.feature_names)
df['target'] = y
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = log.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
y_pred[:10]
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(criterion='gini')
tree.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt 
plt.figure(figsize=(50,50))
plot_tree(tree, feature_names = cancer.feature_names, filled=True)
plt.show()
```

```python
y_pred = tree.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
# entropy
tree = DecisionTreeClassifier(criterion='entropy')
tree.fit(X_train,y_train)

y_pred = tree.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
# entropy
tree = DecisionTreeClassifier(criterion='entropy', max_depth = 3)
tree.fit(X_train,y_train)

y_pred = tree.predict(X_test)
print("Accuracy score is",accuracy_score(y_test,y_pred))

plt.figure(figsize=(50,50))
plot_tree(tree, feature_names = cancer.feature_names, filled=True)
plt.show()
```

```python
# entropy
tree = DecisionTreeClassifier(criterion='gini', max_depth = 3)
tree.fit(X_train,y_train)

y_pred = tree.predict(X_test)
print("Accuracy score is",accuracy_score(y_test,y_pred))

plt.figure(figsize=(50,50))
plot_tree(tree, feature_names = cancer.feature_names, filled=True)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

tree = DecisionTreeClassifier(random_state=0)

param_grid = {'criterion': ["gini",'entropy'],
              'max_depth': [2,3,4,5,6],
              'splitter': ["best", "random"]}

grid = GridSearchCV(estimator = tree, 
                    param_grid = param_grid,
                    cv = 5,
                    scoring='accuracy',
                    verbose=1) # verbose = 1 display the message
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
from sklearn.model_selection import RandomizedSearchCV

randomcv = RandomizedSearchCV(estimator = tree, 
                              param_distributions = param_grid,
                              cv = 5,
                              n_iter = 5,
                              verbose =1)
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_estimator_
```

```python
#post pruning
tree = DecisionTreeClassifier(random_state=0).cost_complexity_pruning_path(X_train,y_train)
ccp_alphas,impurities = tree.ccp_alphas,tree.impurities
```

```python
fig, ax = plt.subplots()
ax.plot(ccp_alphas[:-1], impurities[:-1], marker='o', drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random = RandomForestClassifier(n_estimators=50, 
                                criterion='entropy',
                                max_features = 20,
                                random_state=0,
                                max_depth=4)
```

```python
random.fit(X_train,y_train)
```

```python
y_pred = random.predict(X_test)
accuracy_score(y_test,y_pred)
```

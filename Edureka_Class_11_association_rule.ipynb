# Association Rule Mining

```python
!wget https://www.dropbox.com/s/iivd4novdhlk7lb/groceries.csv
import pandas as pd
import numpy as np
```

```python
fileobj = open("groceries.csv","rt")
content = fileobj.readlines()
fileobj.close()
```

```python
content[:3]
```

```python
'citrus fruit,semi-finished bread,margarine,ready soups\n'.strip() # remove the extra spaces in the string
```

```python
txns = [x.strip() for x in content]
txns[:3]
```

```python
'citrus fruit,semi-finished bread,margarine,ready soups'.split(",")
```

```python
all_txns = [each_txn.split(",") for each_txn in txns]
all_txns[:3] # data in the form of list inside a list
```

```python
# Convert the data into one hot representation
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
one_hot_encoding = TransactionEncoder()
one_hot_encoding.fit(all_txns)
one_hot_txns = one_hot_encoding.transform(all_txns)
```

```python
one_hot_txns.shape
```

```python
one_hot_txns[:2]
```

```python
one_hot_txns = one_hot_txns.astype("int")
one_hot_txns[:2]
```

```python
one_hot_encoding.columns_
```

```python
one_hot_txns_df = pd.DataFrame(one_hot_txns, columns= one_hot_encoding.columns_)
one_hot_txns_df.head()
```

```python
# Frequent itemsets
frequent_itemsets = apriori(one_hot_txns_df, min_support=0.02, use_colnames=True)
frequent_itemsets
```

```python
# build association rules
rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold=0.4)
rules
```

```python
rules.sort_values("lift", ascending=False).head(15)
```

```python
# if lift > 1 --> item A purchased will increase the probability of purchasal of item B
# if lift = 1 --> item A and item B are independent
# if lift < 1 --> item A and item B are alternate products to each other
```
